# Outliers Treatment by IQR method

```python
def iqr(data, col_list):
#     dfn_copy = dfn.copy()
    for col in col_list:
        if np.issubdtype(data[col].dtype, np.number):  # check if data is numeric
            if data[col].dtype == np.float64:  # check if data is decimal
                precision = 2  # get decimal precision
            else:
                precision = 0
            
            Q1 = data[col].quantile(0.25).round(precision)
            Q3 = data[col].quantile(0.75).round(precision)
            IQR = (Q3 - Q1).round(precision)
            LB = (Q1 - 1.5*IQR).round(precision)
            UB = (Q3 + 1.5*IQR).round(precision)
            
            data.loc[data[col] > UB, col] = UB
            data.loc[data[col] < LB, col] = LB
        else:
            continue
    return data
```

```python
def outliers_by_zscore(data, columns):
    for col in columns:
        if data[col].dtype == 'int64' or data[col].dtype == 'float64':
            # calculate the z-score for each column
            z_scores = (data[col] - data[col].mean()) / data[col].std()

            # identify outliers using a threshold of z-score > 3 or z-score < -3
            outliers = np.abs(z_scores) > 3

            if outliers.any():
                # replace outliers with the median value of the column
                data.loc[outliers, col] = data[col].median()
                print(f"Replaced {outliers.sum()} outliers in column '{col}'")

            else:
                print(f"No outliers found in column '{col}'")

        else:
            print(f"Skipping non-numeric column '{col}'")

    return data
```

```python
def draw_emperical(data,columns):
        for i, col in enumerate(columns):
            if data[col].dtype == 'int64' or data[col].dtype == 'float64':
                    mean = data[col].mean()
                    std = data[col].std()
                    fig, axs = plt.subplots(nrows=1, ncols=len(data.columns), figsize=(12, 3))
                    x = np.linspace(data[col].min(), data[col].max(), 1000)
                    pdf = 1/(std * np.sqrt(2 * np.pi)) * np.exp(-((x - mean)**2 / (2 * std**2)))
                    axs[i].plot(x, pdf, color='blue')
                    axs[i].set_xlabel('Values')
                    axs[i].set_ylabel('Probability Density')
                    axs[i].set_title(f"Empirical Standard Deviation Graph for {col}")
```
