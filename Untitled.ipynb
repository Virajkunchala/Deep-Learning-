# Outliers Detection

```python
def finding_outliers(data):
    for x in data.columns:
        if data[x].dtype=='int64' or data[x].dtype=='float64':
            sns.boxplot(data[x])
            plt.show()
            sns.distplot(data[x])
            plt.show()
        else:
            pass
```

# Outliers Treatment by IQR method

```python
dfn_copy = df.copy()
def iqr(dfn, col_list):
#     dfn_copy = dfn.copy()
    for col in col_list:
        if np.issubdtype(dfn_copy[col].dtype, np.number):  # check if data is numeric
            if dfn_copy[col].dtype == np.float64:  # check if data is decimal
                precision = 2  # get decimal precision
            else:
                precision = 0
            
            Q1 = dfn_copy[col].quantile(0.25).round(precision)
            Q3 = dfn_copy[col].quantile(0.75).round(precision)
            IQR = (Q3 - Q1).round(precision)
            LB = (Q1 - 1.5*IQR).round(precision)
            UB = (Q3 + 1.5*IQR).round(precision)
            
            dfn_copy.loc[dfn_copy[col] > UB, col] = UB
            dfn_copy.loc[dfn_copy[col] < LB, col] = LB
        else:
            continue
    return dfn_copy
```

```python
def iqr(dfn,col_list):
    for col in col_list:
#         print(col)
        if dfn[col].dtype!= object:
            Q1 = dfn[col].quantile(0.25)
            Q3 = dfn[col].quantile(0.75)
            IQR = Q3 - Q1
            S = 1.5*IQR
            LB = Q1 - S
            UB = Q3 + S
            dfn.loc[dfn[col] > UB,col] = UB
            dfn.loc[dfn[col] < LB,col] = LB
        else:
            break
    return dfn
```
